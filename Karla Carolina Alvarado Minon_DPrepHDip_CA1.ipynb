###### Drop dataframe columns based on NaN percentage

Can we drop all our missing values? The answer will be "it depends" there are so many factors when dropping missing values might not affect and won't change the result. 

When our data has lots of NaN's there are different things that we can apply instead like dropping columns that have more than 60% of NaNs  they don't bring much information as most of their values are missing.
 
To delete columns based on percentage of NaN values in columns, we can use a pandas dropna() function. It can delete the columns or rows of a dataframe that contains all or few NaN values. As we want to delete the columns that contains either N% or more than N% of NaN values, so we will pass following arguments in it,

axis=1 : Drop columns which contain missing value.
thresh=min_count : Delete columns which contains less than min_count number of non-NaN values.
We calculated this min_count based on percentage of NaN values i.e. N%.


```python
# Delete columns containing either 60% or more than 60% NaN Values
perc = 60.0 #N%
min_count =  int(((100-perc)/100)*aps_failure_pos_df.shape[0] + 1) 
aps_failure_mod= aps_failure_pos_df.dropna( axis=1, 
                thresh=min_count)
```

#### What is PCA?

Principal component analysis (PCA) is a technique that transforms high-dimensions data into lower-dimensions while retaining as much information as possible.

PCA is extremely useful when working with data sets that have a lot of features. While having more data is always great, sometimes they have so much information in them, we would have impossibly long model training time and the curse of dimensionality starts to become a problem. Sometimes, less is more.

A good example for this:

Finding the time to read a 1000-pages book is a luxury that few can afford. Wouldn’t it be nice if we can summarize the most important points in just 2 or 3 pages so that the information is easily digestible even by the busiest person? We may lose some information in the process, but hey, at least we get the big picture.


### Curse of Dimensionality (Bellman,1956) 

When we think about the Curse of Dimensionality, we are talking about dimensions.
Imagine a dataset and its characteristics probably has 20 or 30 independent variables it is quite easy to read. Now think about another dataset, this time it has more than a hundred independent variables and it makes the dataset more complex. This will be known as the curse of dimensionality, the main problem when is about lots of independent variables is that it is difficult to understand us and could be also difficult for the algorithms that we are using which might affect our results.

To understand The Curse of Dimensionality better let's explain the dimensions that we have, as human beings we can see things in 2D and 3D if we have a date set with more variables it will create more dimensionality.

As humans, we can easily analyze 2 or even 5 dimensions but when we have more it is harder to see through the data.

One of the solutions that we can apply in the case of a big dataset is to make sure that the selection of the variables is going to apport valuable information,we don't always need all variables.


To implement PCA it will be necessary to import the following libraries:

Here we are importing the PCA function from sklearn. Note there are a few Python PCA functions to choose from. The one in sklearn is the most commonly used as I know.

The preprocessing package from skleanr gives us functions for scaling the data before performing PCA.

```python
#libraries for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

After import our libraries we are going to drop our first column "class" as it is not a numerical column and PCA only works with numerical data.

```python
#using drop to remove class
newdataframe = aps_final.drop(['class'],axis=1)
```

Perform Scaling on the Data, this means that we need to center and scale the data. To scale our data, we would use StandardScaler which is avaibable in sklearn.

```python
scaler = StandardScaler()
scaler.fit(newdataframe)
aps_final_scaled = scaler.transform(newdataframe)
```

Now we can see our new dataset.

```python
pca_data = pd.DataFrame(aps_final_scaled)
pca_data.head()
```

The new dataset has and std of 1.000008e+00	after scale.

```python
#getting the describe of our new dataset 
pca_data.describe()
```

The goal of thePCA is to ﬁnd the space, which represents the direction of the maximum variance of the given data.

We create a PCA object. Sklearn used objects that can be trained using one dataset and applied to another dataset.

Now **.fit**  is where we do all of the PCA math (i.e. calculate scores and the variation each principal accounts for)
**.transform** this is where we generate coordinates for PCA graph based on the loading scores and the scaled data.

We want to estimate how many components are needed to describe our data. This can be determined by looking at the cumulative explained variance ratio as a function of the number of components:

This curve quantifies how much of the total, 160-dimensional variance is contained within the first N components.

```python
pca = PCA().fit(pca_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('comulative explained variance');
```

We can see that 80, 90, or 100 are account for just over 90% of the variance. That would lead us to believe that using either 90 or 100 components, we would recover most of the essential characteristics of the data.
Let's use 90.

```python
#perfom PCA
pca = PCA(90)
projected = pca.fit_transform(pca_data)
```

The shape attribute returns the dimensions of our data.
In this case we have reduced our independent vairables to 90 so it give us (60000, 90).

Dimensionality reduction is one of the preprocessing steps in many machine learning applications and it is used to transform the features into a lowerdimension space.#

What we are doing here is using our new dataset with only 90 characteristics and it will be easier to apply a machine learning model.

```python
from sklearn.model_selection import train_test_split
X = aps_array[:,0:90]
y = aps_array[:,90]
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.20, random_state=1)
```

We need to import our models and try to see what of those perform better.

```python
from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
```

```python
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
	cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

```python
pyplot.boxplot(results, labels=names)
pyplot.title('Algorithm Comparison')
pyplot.show()
```

This graph help us to visualize all models and see which one works better in this case I would pick LDA as the box plot looks less skewed.


```python
model = LogisticRegression(solver='liblinear', multi_class='ovr')
model.fit(X_train, Y_train)
predictions = model.predict(X_validation)
```

```python
print(accuracy_score(Y_validation, predictions))
print(confusion_matrix(Y_validation, predictions))
```

We have an accuracy of 0.99 that means that our algorithm is working well.
